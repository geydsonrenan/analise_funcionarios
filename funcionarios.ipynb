## Visualização e Modificação dos Dados

```python
#transformando a coluna Attrition em binário para facilitar a análise
func_df['Attrition'] = func_df['Attrition'].astype("category")
func_df['Attrition'] = func_df['Attrition'].cat.codes
func_df.head()
```

```python
func_df['OverTime'] = func_df['OverTime'].astype("category")
func_df['OverTime'] = func_df['OverTime'].cat.codes
func_df['OverTime'].head()
```

```python
func_df['Over18'] = func_df['Over18'].astype("category")
func_df['Over18'] = func_df['Over18'].cat.codes
func_df['Over18'].head()
```

```python
func_df.hist(bins=30, figsize=(20,20), color='b');
```

EmployeeCount, StandardHours e Over18 possuem apenas um mesmo valor em todas as linhas, portanto devem ser apagadas

```python
func_df.drop(['EmployeeCount', 'StandardHours', "Over18", 'EmployeeNumber'], inplace=True, axis=1)
```

```python
func_df.shape
```

```python
left_df = func_df[func_df['Attrition'] == 1]
stayed_df = func_df[func_df['Attrition'] == 0]
```

```python
print('Total = ', len(func_df))
print('Número de funcionários que saíram da empresa = ', len(left_df))
print('Porcentagem de funcionários que saíram da empresa = ', (len(left_df) / len(func_df)) * 100)
print('Número de funcionários que ficaram na empresa = ', len(stayed_df))
print('Porcentagem de funcionários que ficaram na empresa = ', (len(stayed_df) / len(func_df)) * 100)
```

```python
left_df.describe()
```

```python
stayed_df.describe()
```

As estatísticas acima indicam que as pessoas mais novas estão mais propensas a sair da empresa pela diferença entre a média da idade. Também há uma diferença entre o salário recebido, sendo o salário das pessoas que saíram mais baixo, levando a entender que os funcionários podem ter saído para procurar um emprego com um salário mais atrativo. Outro ponto importante é sobre a satisfação do funcionário com a sua função na empresa, que é bem menor evidenciada nos funcionários que saíram, indicando a lógica de insatisfação ser um gatilho para o abandono do cargo.

## Pré-Processamento e Divisão dos Dados

```python
func_df.head()
```

```python
X_cat = func_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
X_cat.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat = pd.DataFrame(X_cat)
```

```python
X_cat
```

```python
X_numerical = func_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
X_numerical
```

```python
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_all.columns = X_all.columns.astype(str)
X = scaler.fit_transform(X_all)
```

```python
y = func_df['Attrition']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape
```

## Regressão Logística

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
```

```python
prediction_y = logistic_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction_y)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, prediction_y)
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, prediction_y)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, prediction_y)
```

```python
from sklearn.metrics import f1_score, classification_report
f1_score(y_test, prediction_y, average='macro')
```

```python
print(classification_report(y_test, prediction_y))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
prediction_y = rf.predict(X_test)
```

```python
accuracy_score(y_test, prediction_y)
```

```python
cm = confusion_matrix(y_test, prediction_y)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, prediction_y))
```

## Redes Neurais

```python
import tensorflow as tf
rn = tf.keras.models.Sequential()
# foi colocado 25 nós por termos 50 atributos nos dados e apenas uma saída
rn = tf.keras.models.Sequential()
rn.add(tf.keras.layers.Dense(units = 25, activation='relu', input_shape=(50,)))
rn.add(tf.keras.layers.Dense(units = 25, activation = 'relu'))
rn.add(tf.keras.layers.Dense(units = 25, activation = 'relu'))
rn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
rn.summary()
```

```python
rn.compile(optimizer="Adam", loss='binary_crossentropy', metrics=['accuracy'])
```

```python
rn.fit(X_train, y_train, epochs=200)
```

```python
prediction_y = rn.predict(X_test)
```

```python
prediction_y = (prediction_y >= 0.5)
prediction_y
```

```python
cm = confusion_matrix(y_test, prediction_y)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, prediction_y))
```
